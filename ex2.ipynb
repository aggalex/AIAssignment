```python
import sentiment
from sentiment import *
import pandas as pd

cls = TwitterClassifier(correction=False)
cls.words
```

```python
data = pd.read_csv('apple-twitter-sentiment-texts.csv')
data
```

## Data Preprocessing

We format the data in the model that our bayesian classifier supports

```python
data['sentiment'] = data['sentiment'].map(lambda n:
                                          Sentiment.NEGATIVE if n < 0 else
                                          Sentiment.POSITIVE if n > 0 else
                                          Sentiment.NEUTRAL)
data.rename(columns={'text':'tweet'}, inplace=True)
data
```

And then we split into a training and a testing dataset

```python
n = data.shape[0]
n_train = (n * 7) // 10         # 70% of the data goes for training
train_data = data.iloc[n_train:, :]
test_data = data.iloc[:n_train, :]
```

## Training

We train our model based on our training data

```python
cls.train(train_data)
```

Those are all the keywords the model has found

```python
cls.words
```

## Testing

We test the model's accuracy based on the testing data

```python
cls.test(test_data)
```
